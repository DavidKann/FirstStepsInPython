If you have a single item of data to store you can make a variable

When you have many, related items you can make a list:

```python
students = ["Andy", "Susan", "Simon"]
print(students)
```

Lists can be changed, dynamically:

```python
students = ["Andy", "Susan", "Simon"]
new_student = await input("Enter the new student's name")
```

```python
students.append(new_student)
print(students)
```

You can access a single item from the list using its index

```python
students = ["Andy", "Susan", "Simon"]
print(students[1])
```

Which student was printed when you called for students[1]? Did you expect it to be Andy?

You can remove items from a list:

```python
students = ["Andy", "Susan", "Simon"]
print(students)
remove_student = await input("Which student to remove?")
```

```python
students.remove(remove_student)
print(students)
```

You can ask a list to sort itself:

```python
grades = [90,85,77,89,91]
grades.sort()
print("The top score was", grades[len(grades)-1])
```

You can combine data in a list:

```python
grades = [90,85,77,89,91]
sum = 0
for grade in grades:
    sum = sum + grade

average = sum / len(grades)
print("The average grade is: ", average)
```

You can use the index to work across many lists:

```python
students = ["Andy", "Susan", "Simon", "Peter", "Harold"]
grades = [90,85,77,89,91]
for i in range(len(students)):
    print(students[i] + "'s grade is ", grades[i])
```

The data can also be combined into a single list of pairs of data:

```python
student_data = [["Andy",90],["Susan",85],["Simon",77],["Peter",89],["Harold",91]]

for student in student_data:
    print(student[0] + "'s grade is ", student[1])
    
top_score = student_data[0][1]
top_student = ""
for student in student_data:
    if student[1] > top_score:
        top_score = student[1]
        top_student = student[0]
 
print(top_student, "scored the top score:", top_score)
```

- But you can also use a function to describe your own type of sorting order:

```python
student_data = [["Andy",90],["Susan",85],["Simon",77],["Peter",89],["Harold",91]]

def alphabetizer(student):
    return student[0]

student_data.sort(key = alphabetizer)
print(student_data)
```
